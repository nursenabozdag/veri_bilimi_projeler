### Geçen Ortalama Sürenin Test Edilmesi

Elimizde bir web sitesinde geçirilen ortalama süre gerçekten 170 sn mi? sorusu var. Bu süreye göre reklam veya içerik araştırılması yapılacaktır. Bundan dolayı bu sürenn istatistiğini hesaplamak istiyoruz.
Hipotezler:

$H_0: \mu=170$

$H_1: \mu !=170$

```python
#kütüphanemizi import ediyoruz.
import numpy as np
```

```python
#elimizdeki ölçümleri numpy array olarak kaydediyoruz.

olcumler = np.array([17,160,345,149,145,78,90,201,225,211,119,157,234,134,156,177,177,253,123,145,67,54,78,99,155,84,224,11,151,124,456,63])
```

```python
#ilk 10 gözlemi görelim.
olcumler[0:10]
```

```python
#istatistik için kullanacağımız scipy.stats kütüphanesini import edelim.
import scipy.stats as stats
```

```python
stats.describe(olcumler)
```

```python
#elimizdeki verinin normal olduğunu varsaylım ve histogram grafiği ile bunu gösterelim.
import pandas as pd
pd.DataFrame(olcumler).plot.hist();
```

```python
#Elimzde bulunan hipotezin normallik varsayımı için kullanılan diğer bir teknik qqplot tekniğidir. Bu teknik için pylab kütüphanesinden yararlanacağız.
import pylab
stats.probplot(olcumler, dist  ="norm", plot =pylab)
pylab.show()
```

```python
#Diğer bir teknik ise Shapiro-Wilks testidir.
from scipy.stats import shapiro
shapiro(olcumler)
```

```python
#shapiro testini kullanarak başka bir şekilde gösterimi;
print("Test İstatistiği:" + str(shapiro(olcumler)[0]))
print("Hesaplanan P-value:" + str(shapiro(olcumler)[1]))
```

```python
#hipotez testtini uygulayalım;
stats.ttest_1samp(olcumler, popmean =170)
```

Bizim hipotezimiz ortalama 170 sn üzerine kuruluydu. Sonuç olarak biz H0 hipotezini redderiz, çünkü ortalama geçirilen süre 170 sn'den küçüktür.
